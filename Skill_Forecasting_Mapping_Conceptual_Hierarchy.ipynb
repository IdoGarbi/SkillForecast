# Dataset Definition

Secondly, we need to define the lexical dataset


```
!pip install nltk

import nltk
nltk.download('wordnet')
from nltk.corpus import wordnet

lexical_dictionary = {}

word_counter=0
word_maximum= 70000 
for concept_array in wordnet.all_synsets():


  w_name= concept_array.name().split(".")[0].replace("_"," ")
  w_def= str(concept_array.definition())
  lexical_dictionary[w_name] = w_def

  word_counter=word_counter+1
  if (word_counter == word_maximum):
    break

print(lexical_dictionary)

syns = wordnet.synsets(str(examined_skill))

if (len(syns) > 0):
  print("Testing... Definition of \"" +examined_skill+"\"--> "+syns[0].definition())
  print("--> Success! Stage two defined!")
else:
  print("Skill not found in dataset")

```

# Optional - Adjustment for Current Automation Level

We need to discard previously automated concepts. In this work, it is a mockup and not actual data, but the discarding mechanism is the same

```
# (i) Generate the previously automated concepts- mockup
previously_automated_concepts = {}

discarded_concepts_counter=0
concepts_to_discard= 10000 
for concept_array in wordnet.all_synsets():

  w_name= concept_array.name().split(".")[0].replace("_"," ")
  w_def= str(concept_array.definition())
  previously_automated_concepts[w_name] = w_def

  discarded_concepts_counter=discarded_concepts_counter+1
  if (discarded_concepts_counter == concepts_to_discard):
    break

# (i) remove them from the concept dataset
for concept_array in previously_automated_concepts:
   lexical_dictionary[concept_array]="" # (!) reduces abstraction level to zero
   
print("Success, previously automated concepts discarded!")
```

# Function Definition

Thirdly, we need to define the functions we intend to use

```
# Print Title
def print_title(title):
  print()
  print("///////////////////////////////////////")
  print(title)
  print("///////////////////////////////////////")
  print()

# Recursion function - Build concept hierarchy, i.e. how many concepts are needed to define your concept
def build_concept_hierarachy(word, lexical_data, learned_concepts,maximal_abstraction_level,depth_level,compound_abstraction_level):
  
  # (i) Define key variables
  recursion_max_depth=900 # (i) For resource consumption optimization, limit maximal recursion depth 
  
  concept_known= False
  concept_found= False

  printing= False

  if len(maximal_abstraction_level) <= depth_level: # if new abstraction level peaked
    maximal_abstraction_level.append("Level "+str(depth_level))

  white_space=" "
  adjusted_depth_level= depth_level*3
  for x in range(adjusted_depth_level):
    white_space = white_space+" "

  if printing is True:
    print(white_space+" "+word)

  # (i) Look for the word in lexical data
  # //////////////////////////////////////////////////////
  word_def_data= lexical_dictionary.get(word) 

  if word_def_data is not None and len(word_def_data)>0:
    concept_found= True
    word_lexical_definition=word_def_data
    if printing is True:
      print(white_space+"   --> Definition- "+str(word_lexical_definition))
  # //////////////////////////////////////////////////////

  # (i) For each word, check if you already "know" it, and if so, remove it
  if learned_concepts.get(word) is None:
    # (i) Add to learned_concepts
  
    if concept_found is True:
      learned_concepts[word] = "Found"

    else:
      learned_concepts[word] = ""
      if printing is True:
        print(white_space+"   --> Couldn't find "+word)
    
  else:
    concept_known= True
    if printing is True:
     print(white_space+"   --> I already know "+word)

  # ////////////////////////////////////////////////////////////////////////////////////////
  # (i) If not, explore the hierarchy via recursion
  if concept_found is True and concept_known is False:
    if printing is True:
      print(white_space+"   --> must go deeper")
    # /// Recursion process marker


    lexical_definition_components = word_lexical_definition.split()

    new_depth_level= depth_level+1

    if (new_depth_level < recursion_max_depth): # if the new depth level is smaller than the depth limit
      for component in lexical_definition_components:
        build_concept_hierarachy(component, lexical_data, learned_concepts,maximal_abstraction_level,new_depth_level,compound_abstraction_level)
    else:
      compound_abstraction_item={word,depth_level}
      compound_abstraction_level.append(compound_abstraction_item)

  #////////////////////////////////////////////////////////////////////////////////////////
    

def build_concept_reliance(examined_skill,lexical_dictionary,linked_concepts,maximal_reliance_level,depth_level,compound_reliance_level):

  recursion_max_depth=2 # (i) For resource consumption optimization, limit maximal recursion depth 
  printing=False

  white_space=" "
  adjusted_depth_level= depth_level*3
  for x in range(adjusted_depth_level):
    white_space = white_space+" "

  if print is True:
    print()
    print(white_space+ "Depth Level- "+str(depth_level))

  # (i) Search the lexical dictionary values for any item that contains the concept
  for concept,lexical_value in lexical_dictionary.items():
    concept_linked=False

    if examined_skill in lexical_value:
      if print is True:
        print(white_space+"!--!")
        print(white_space+concept+"---"+lexical_value)
        print(white_space+"!--!")
        print()


      # (i) check if we already "relied" on this item- if so, there is no need to re-check it
      if linked_concepts.get(concept) is not None:
        if print is True:
          print(concept+" Already Found!!!")
        concept_linked=True
      else:
        # insert the concept to linked concepts
        linked_concepts[concept] = "Linked"


      # (i) Recursion gateway
      if concept_linked is False:
        # (i) edit reliance level  
        if len(maximal_reliance_level) <= depth_level: # if reliance level peaked
          maximal_reliance_level.append("Reliance Level "+str(depth_level))
        
        # (i) explore further via recursive function
        new_depth_level=depth_level+1
        if (new_depth_level < recursion_max_depth): # if the new depth level is smaller than the depth limit
          build_concept_reliance(concept,lexical_dictionary,linked_concepts,maximal_reliance_level,new_depth_level,compound_reliance_level)
        else:
          compound_reliance_item={concept,depth_level}
          compound_reliance_level.append(compound_reliance_item)

  if print is True:
    print()



print("--> Functions set- stage three defined!")
```
